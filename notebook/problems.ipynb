## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

def roll_dice():
    return random.randint(1, 6)

def simulate_experiment(num_simulations):
    count_even_or_greater_than_7 = 0

    for _ in range(num_simulations):
        # Roll two dice
        dice1 = roll_dice()
        dice2 = roll_dice()

        # Calculate the sum
        total = dice1 + dice2

        # Check conditions
        if total > 7 or total % 2 == 0:
            count_even_or_greater_than_7 += 1

    probability_even_or_greater_than_7 = count_even_or_greater_than_7 / num_simulations

    return probability_even_or_greater_than_7

# Run the simulation 1000 times
num_simulations = 1000
prob_even_or_greater_than_7 = simulate_experiment(num_simulations)

print(f"Estimated probability of getting an even number or a sum greater than 7: {(prob_even_or_greater_than_7)*100}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random

# Define the number of trials
num_trials = 1000

# Define the number of balls of each color
num_white = 10
num_red = 20
num_green = 30

# Total number of balls
total_balls = num_white + num_red + num_green

# Define the probabilities of each color
prob_white = num_white / total_balls
prob_red = num_red / total_balls
prob_green = num_green / total_balls

# Function to simulate the experiment
def simulate_experiment():
    # Draw 5 balls with replacement
    balls_drawn = random.choices(['white', 'red', 'green'], k=5)

    # Check for the events
    event_3w_2r = balls_drawn.count('white') == 3 and balls_drawn.count('red') == 2
    event_all_same_color = len(set(balls_drawn)) == 1

    return event_3w_2r, event_all_same_color

# Initialize counters for events
count_3w_2r = 0
count_all_same_color = 0

# Run the experiment for the specified number of trials
for _ in range(num_trials):
    event_3w_2r, event_all_same_color = simulate_experiment()

    # Increment counters if events occur
    count_3w_2r += event_3w_2r
    count_all_same_color += event_all_same_color

# Calculate probabilities
prob_3w_2r = count_3w_2r / num_trials
prob_all_same_color = count_all_same_color / num_trials

# Print the results
print(f"Probability of taking 3 white and 2 red: {prob_3w_2r}")
print(f"Probability of all balls being the same color: {prob_all_same_color}")
```
